Convertendo Datas

```python

column_date = ['order_date', 'delivery_date']
for column in column_date:
    df[column] = pd.to_datetime(df[column], errors='coerce')
```

Tratando Nulos

```python
df['coupon_code'] = df['coupon_code'].fillna('Sem cupom')
df['reason_for_cancellation'] = df['reason_for_cancellation'].fillna('Pedido não foi cancelado')
df['delivery_date'] = df['delivery_date'].fillna('Pedido não entregue')
```
